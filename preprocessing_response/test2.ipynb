```python
import json
# load the json file
with open('actual_res2.json',encoding='utf-8') as f:
    data = json.load(f)

data = data['textAnnotations']
data
```

```python
def get_x_and_y(data,word):
    for i in range(len(data)):
        if (data[i]['description'] == word):
            print(data[i])
            return data[i]['boundingPoly']['vertices'][0]['x'],data[i]['boundingPoly']['vertices'][0]['y'],i
    return 0,0,0
```

```python
stud_x,stud_y,index = get_x_and_y(data,'student')
```
