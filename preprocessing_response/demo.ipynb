```python
import xlwt
```

```python
def make_excel(my_dict):
    wb = xlwt.Workbook()
    ws = wb.add_sheet('sheet1')
    ws.write(0,0,"Roll No")
    ws.write(0,1,"Name")
    ws.write(0,2,"MCQ Marks")
    ws.write(0,3,"Q2 Marks")
    ws.write(0,4,"Q3 Marks")
    ws.write(0,5,"Total Descriptive Marks")
    ws.write(0,6,"Total Marks")
    row = 6
    for key,value in my_dict.items():
        ws.write(row,0,key)
        ws.write(row,1,value[0]["Name"])
        ws.write(row,2,value[0]["MCQ Marks"])
        ws.write(row,3,value[0]["Q2 Marks"])
        ws.write(row,4,value[0]["Q3 Marks"])
        q_2_marks = int(value[0]["Q2 Marks"])
        q_3_marks = int(value[0]["Q3 Marks"])
        mcq_marks = int(value[0]["MCQ Marks"])
        total_descriptive_marks = q_2_marks + q_3_marks
        total_marks = mcq_marks + total_descriptive_marks
        ws.write(row,5,total_descriptive_marks)
        ws.write(row,6,total_marks)
        row += 1
    wb.save('result.xls')
```

```python
make_excel(data)
```
